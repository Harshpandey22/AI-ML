**Encoding**

Encoding is conversion of objects into int or float (i.e, strings/categorical values into int/float)

Types of Encoding:
1. One hot encoding
2. Label encoding
3. Manual encoding

Need for encoding is to pass data to algoritm for model training as it will only accept numerical data not categorical data

```python
# One Hot Encoding

pd.get_dummies(df['model'])       #get.dummies is used to encode categorical into numerical, but is not used frequently

# 0 means the model name is not present at that index whereas 1 means that name is present in pertical column for a perticular index
```

```python
# Label Encoding
# need to use sklearn library and import LabelEncoder

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()     #initializing properties of LabelEncoder to a variable
```

```python
df3['gender'] = le.fit_transform(df3['gender'])
df3
```

```python
df['model'] = le.fit_transform(df['model'])
df
```
