```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# Encoding Categorical Data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.State = le.fit_transform(df.State)
```

```python
df.head()
```

```python
# X and Y Split
X = df.drop(columns=['Profit'],axis=1)
X.head()
```

```python
y=df.Profit
y.head()
```

```python
# Scaling the DATA
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
```

```python
X_Scaled = pd.DataFrame(scale.fit_transform(X),columns=X.columns)
X_Scaled.head()
```

```python
# TRAIN AND TEST SPLIT
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_Scaled,y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Model Building

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
r = Ridge()
l = Lasso()
```

```python
r.fit(X_train,y_train)
```

```python
l.fit(X_train,y_train)
```

```python
pred1 = r.predict(X_test)
pred1
```

```python
# Traning Accuracy
pred1_train = r.predict(X_train)
```

```python
pred2 = l.predict(X_test)
pred2
```

```python
# Traning Accuracy
pred2_train = l.predict(X_train)
```

```python
profit=pd.DataFrame(({'Actual Profit':y_test, 'Ridge_test':pred1, 'Lasso_test':pred2}))
profit.head(10)
```

### Evaluating Ridge and Lasso Models

```python
from sklearn import metrics
```

```python
# R2 Score
print(metrics.r2_score(y_test,pred1))
print(metrics.r2_score(y_test,pred2))
```

```python
# MSE
print(metrics.mean_squared_error(y_test,pred1))
print(metrics.mean_squared_error(y_test,pred2))
```

```python
# RMSE
print(np.sqrt(metrics.mean_squared_error(y_test,pred1)))
print(np.sqrt(metrics.mean_squared_error(y_test,pred2)))
```
